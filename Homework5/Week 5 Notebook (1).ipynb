```python
# Python 3 program to find
# factorial of given number
def factorial(n):

    # Checking the number
    # is 1 or 0 then
    # return 1
    # other wise return
    # factorial
    if (n==1 or n==0): #base case
        return 1
    else:
        return (n * factorial(n - 1))

# Driver Code
num = 100;
print("number : ",num)
print("Factorial : ",factorial(num))
```

```python
# Recursive Python function to solve tower of hanoi


def TowerOfHanoi(n, from_rod, to_rod, aux_rod):
    if n == 0:
        return
    TowerOfHanoi(n-1, from_rod, aux_rod, to_rod)
    print("Move disk", n, "from rod", from_rod, "to rod", to_rod)
    TowerOfHanoi(n-1, aux_rod, to_rod, from_rod)


# Driver code
N = 3

# A, C, B are the name of rods
TowerOfHanoi(N, 'A', 'C', 'B')

```

```python
import csv
import networkx as nx
import matplotlib.pyplot as plt

# Define a class for the binary search tree node
class TreeNode:
    def __init__(self, purchase_amount, customer_id, product_id, purchase_date):
        self.purchase_amount = purchase_amount
        self.customer_id = customer_id
        self.product_id = product_id
        self.purchase_date = purchase_date
        self.left = None
        self.right = None

# Function to insert a node into the binary search tree
def insert(root, purchase_amount, customer_id, product_id, purchase_date):
    if root is None:
        return TreeNode(purchase_amount, customer_id, product_id, purchase_date)
    
    if purchase_amount < root.purchase_amount:
        root.left = insert(root.left, purchase_amount, customer_id, product_id, purchase_date)
    else:
        root.right = insert(root.right, purchase_amount, customer_id, product_id, purchase_date)
    
    return root

# In-order traversal to calculate total purchase amount
def calculate_total_purchase_amount(root):
    if root is None:
        return 0
    return calculate_total_purchase_amount(root.left) + root.purchase_amount + calculate_total_purchase_amount(root.right)

# In-order traversal to find and display customers with a purchase amount greater than threshold
def find_customers_above_threshold(root, threshold):
    if root is None:
        return
    
    find_customers_above_threshold(root.left, threshold)
    
    if root.purchase_amount > threshold:
        print(f"Customer ID: {root.customer_id}, Product ID: {root.product_id}, Purchase Date: {root.purchase_date}, Purchase Amount: {root.purchase_amount}")
    
    find_customers_above_threshold(root.right, threshold)



# Read and parse the CSV dataset
def read_and_parse_csv(filename):
    data = []
    with open(filename, 'r') as csvfile:
        csvreader = csv.reader(csvfile)
        next(csvreader)  # Skip the header row
        for row in csvreader:
            customer_id, product_id, purchase_date, purchase_amount = row
            data.append((float(purchase_amount), customer_id, product_id, purchase_date))
    return data

# Function to create a networkx graph from the binary search tree
def build_tree_graph(root):
    G = nx.DiGraph()
    
    def add_nodes_and_edges(node):
        if node:
            G.add_node(node.purchase_amount, label=f"Customer {node.customer_id}\n${node.purchase_amount:.2f}")
            if node.left:
                G.add_edge(node.purchase_amount, node.left.purchase_amount)
                add_nodes_and_edges(node.left)
            if node.right:
                G.add_edge(node.purchase_amount, node.right.purchase_amount)
                add_nodes_and_edges(node.right)
    
    add_nodes_and_edges(root)
    return G



# Main function
def main():
    filename = "C:\\Users\matth\OneDrive\Documents\SJSU\Data 200\purchase_data.csv"   # Change to your dataset filename
    threshold = float(input("Enter the purchase amount threshold: "))

    # Initialize the root of the binary search tree
    root = None

    # Read and parse the CSV dataset
    dataset = read_and_parse_csv(filename)

    # Build the binary search tree
    for purchase_amount, customer_id, product_id, purchase_date in dataset:
        root = insert(root, purchase_amount, customer_id, product_id, purchase_date)

    # Calculate and display the total purchase amount
    total_purchase_amount = calculate_total_purchase_amount(root)
    print(f"Total Purchase Amount: ${total_purchase_amount:.2f}")

    # Find and display customers with a purchase amount above the threshold
    print(f"Customers with Purchase Amount > ${threshold:.2f}:")
    find_customers_above_threshold(root, threshold)
# Create a networkx graph from the binary search tree
tree_graph = build_tree_graph(root)

# Draw the tree using matplotlib
plt.figure(figsize=(10, 5))
pos = nx.spring_layout(tree_graph)  # Position nodes in a tree-like layout
labels = nx.get_node_attributes(tree_graph, 'label')
nx.draw_network(tree_graph, pos, labels=labels, with_labels=True, node_size=5000, node_color='lightblue', font_size=8)
plt.title("Binary Search Tree Visualization")
plt.show()

if __name__ == "__main__":
    main()
```
